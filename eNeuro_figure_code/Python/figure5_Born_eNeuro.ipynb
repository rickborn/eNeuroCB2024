```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a categorical data type using pandas
x = pd.Categorical(['A', 'B', 'C', 'D', 'E'], categories=['A', 'B', 'C', 'D', 'E'], ordered=True)

# Mean ranks for bright and dull groups
# From Table 2, p. 67 of Rosenthal & Lawson, 1964
bright_mean_rank2 = [4.3, 4.9, 5.1, 3.7, 4.1]
dull_mean_rank2 = [5.3, 6.5, 5.8, 4.6, 6.0]

# Create the bar plot
fig, ax = plt.subplots()

# Plot data
bar_width = 0.35
index = range(len(x))

ax.bar(index, bright_mean_rank2, bar_width, label='Bright')
ax.bar([i + bar_width for i in index], dull_mean_rank2, bar_width, label='Dull')

# Set labels, title, and legend
ax.set_xlabel('Lab group')
ax.set_ylabel('Normalized mean rank')
ax.legend()
_ = ax.set_title('Rosenthal & Lawson 1964, Table 2')
```
