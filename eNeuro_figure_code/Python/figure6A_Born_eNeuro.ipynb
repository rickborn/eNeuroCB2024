```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# make a plot of the binomial probabilities of assigning different numbers
# of animals to treatment group
n = 20
p = 0.5

# Compute binomial probabilities
x = np.arange(0, n + 1)
y = binom.pmf(x, n, p)

# Plot all probabilities
plt.bar(x, y, label="All assignments",color='darkblue')

# Color in the bars corresponding to 8:12
y_subset = binom.pmf(np.arange(8, 13), n, p)
plt.bar(np.arange(8, 13), y_subset, label="8 to 12 subjects", color='darkred')

# Add labels
plt.xlabel('# of subjects in placebo group')
plt.ylabel('Probability')
plt.legend()

# Show plot
plt.show()
```

```python
# Probability of getting exactly 10 animals in each group
p10 = binom.pmf(10, n, p)
print(f'Probability of getting exactly 10 animals in each group: {p10}')
```

```python
# Probability of tolerating unevenness: at least 8 animals in each group
p_success = binom.cdf(12, n, p) - binom.cdf(7, n, p)
print(f'Success rate with at least 8 animals in each group: {p_success}')

# Calculate failures directly (probability of fewer than 8 or more than 12 animals)
p_fail = binom.cdf(7, n, p) * 2
print(f'Failure rate (fewer than 8 or more than 12 animals): {p_fail}')
```
